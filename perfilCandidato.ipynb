```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs = {
    'perfil_eleitorado': None
}

def make_df():
    '''Pega uma lista de arquivos e retorna um único dataframe gerado.'''
    # Lista de dataframes
    dfs = [pd.read_csv('perfil_eleitorado/perfil_eleitorado_{}.txt'.format(ano), sep=';',encoding='latin-1') for ano in range(2002, 2017, 2)]
    # Concatena todos os dataframes em um só
    return pd.concat(dfs, ignore_index=True, sort=False)
    #return pd.concat([dfs, pd.read_csv('perfil_eleitorado/perfil_eleitorado_2018.csv',sep=';',encoding='latin-1')])

# Junta todos os dados de perfil eleitorado em um unico dataframe
dfs['perfil_eleitorado'] = make_df()
```
